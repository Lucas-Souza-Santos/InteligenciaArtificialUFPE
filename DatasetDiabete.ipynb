```
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt
```

```
import sklearn
```

```
sklearn.__version__
```

```
x, y = datasets.load_diabetes(return_X_y=True)
```

```
x.shape
```

```
y.shape
```

```
# Correlação
for i in range(0, 10):
  plt.figure(figsize=(8, 6))
  plt.title(f"gráfico {i + 1}")
  plt.scatter(x[:, i], y)
 
```

```
# Correlação de Cada Curva:
print("============= CORRELAÇÃO ============== ")
for i in range(0, 10):
  print(f"Gráfico {i}: {round(np.corrcoef(x[:, i], y)[0][1], 2)}")
```
