# Gráfico de dispersão Matplotlib

**Um gráfico de dispersão é um tipo de gráfico que mostra os dados como uma coleção de pontos** 

A posição de um ponto depende de seu valor bidimensional, onde cada valor é um a posição na dimensão horizontal ou vertical.

```
import numpy as np 
import matplotlib.pyplot as plt

# Create data

N = 500
colors = (0, 0, 0)
x = np.random.rand(N)
y = np.random.rand(N)
area = np.pi*3

plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.title('Gráfico de Dispersão')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

**Gráfico de dispersão com grupos Os**

```
# Create data
N = 60
g1 = (0.6 + 0.6 * np.random.rand(N), np.random.rand(N))
g2 = (0.4 + 0.3 * np.random.rand(N), 0.5 * np.random.rand(N))
g3 = (0.3 * np.random.rand(N), 0.3 * np.random.rand(N))

data = (g1, g2, g3)
colors = ("red", "green", "blue")
groups = ("Coffee", "Tea", "Water")

# Create plot
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

for data, color, group in zip(data, colors, groups):
  x, y = data
  ax.scatter(x,y, alpha=0.8, c=color, edgecolors='none', s=30, label=group)

plt.title('Matplot scatter plot')
plt.legend(loc=2)
plt.show()
```
